# Cmn files 
1) open the file one by one 
2) drop columns named MJdate,Time,PRN,Az,Ele,Stec,S4 
3) take maximum value of Vtec 
4) use geopandas and  shapely.geometry import Point to get Geographic Centroid Latitude,lonitude for the column Lat,Lon 
5) add a column['datetime'] contain 2024-01-04 as  pd.to_datetime().dt.date
6) save all files in dateframe arranged by the datetime
output should by a dateframe contain 819 rows and 3 columns 

```python
import pandas as pd
import os
import geopandas as gpd
from shapely.geometry import Point
from astropy.time import Time

# Define the directory containing your .Cmn files
directory = "AIData"

# Initialize an empty list to store the processed data
data_list = []

# Loop through each file in the directory
for filename in tqdm(os.listdir(directory)):
    if filename.endswith(".Cmn"):
        # Construct the full file path
        file_path = os.path.join(directory, filename)
        
        # Load the file into a DataFrame, skipping the first 3 rows if needed
        df = pd.read_csv(file_path, delim_whitespace=True, skiprows=3)  # Adjust delimiter if needed
        
        # Drop unnecessary columns
        df = df.drop(columns=[ 'Time', 'PRN', 'Az', 'Ele', 'Stec', 'S4'])
        
        # Take the maximum value of Vtec
        max_vtec = df['Vtec'].max()
        
        # Calculate the geographic centroid
        points = [Point(xy) for xy in zip(df['Lon'], df['Lat'])]
        gdf = gpd.GeoDataFrame(df, geometry=points)
        centroid = gdf.geometry.unary_union.centroid
        
        # Convert MJdate to a modern datetime format
        df['datetime'] = pd.to_datetime(Time(df['MJdatet'], format='mjd').to_datetime()).date
        
        # Create a new DataFrame row with the needed information
        new_row = {
            'datetime': df['datetime'].iloc[0],  # Assuming all rows have the same date
            'max_vtec': max_vtec,
            'centroid_lat': centroid.y,
            'centroid_lon': centroid.x
        }
        
        # Append the new row to the list
        data_list.append(new_row)

# Convert the list of dictionaries to a DataFrame
final_df = pd.DataFrame(data_list)
final_df 
```

```python
# Ensure the datetime column is in datetime format
final_df['datetime'] = pd.to_datetime(final_df['datetime'])

# Sort the DataFrame by the datetime column in ascending order
final_df = final_df.sort_values(by='datetime').reset_index(drop=True)
final_df
```
